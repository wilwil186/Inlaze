```python
import os
import pandas as pd

# Directorios
input_folder = "./data/"
output_folder = "./dataexcel/"

# Crear carpeta de salida si no existe
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Obtener lista de archivos CSV en la carpeta de entrada
csv_files = [f for f in os.listdir(input_folder) if f.endswith('.csv')]

# Verificar si hay archivos CSV
if not csv_files:
    print("No se encontraron archivos CSV en la carpeta ./data/")
else:
    for csv_file in csv_files:
        file_path = os.path.join(input_folder, csv_file)  # Ruta del CSV
        df = pd.read_csv(file_path)  # Leer CSV
        excel_file = os.path.join(output_folder, csv_file.replace('.csv', '.xlsx'))  # Nombre de salida

        # Guardar en Excel
        df.to_excel(excel_file, index=False, engine='openpyxl')

        print(f"Archivo guardado: {excel_file}")

print("Proceso completado.")
```
